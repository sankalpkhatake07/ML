```python
x=df.drop(columns=["Satisfaction","ID"],axis=1)
```

```python
x
```

```python
y=df["Satisfaction"]
y
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=5)
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
xgb_model.fit(x_train, y_train)
```

```python
y_pred = xgb_model.predict(x_test)
y_prob = xgb_model.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
print(" Accuracy:", accuracy_score(y_test, y_pred))
print(" ROC-AUC:", roc_auc_score(y_test, y_prob))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm')
plt.title("Confusion Matrix - XGBoost")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'XGBoost (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC–AUC Curve - XGBoost")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

feature_prompts = [
    ('Age: ', int),
    ('Flight Distance: ', int),
    ('Departure Delay (min): ', int),
    ('Arrival Delay (min): ', int),
    ('Class (0=Eco, 1=Eco Plus, 2=Business): ', int),
    ('Check-in Service (1-5): ', int),
    ('In-flight Service (1-5): ', int),
    ('Seat Comfort (1-5): ', int),
    ('Food and Drink (1-5): ', int),
    ('Baggage Handling (1-5): ', int)
]

user_data = {}
for prompt, typ in feature_prompts:
    value = input(prompt)
    user_data[prompt.split('(')[0].strip()] = typ(value)

columns = [
    'Gender', 'Age', 'Customer Type', 'Type of Travel', 'Class',
    'Flight Distance', 'Departure Delay', 'Arrival Delay',
    'Departure and Arrival Time Convenience', 'Ease of Online Booking',
    'Check-in Service', 'Online Boarding', 'Gate Location', 'On-board Service',
    'Seat Comfort', 'Leg Room Service', 'Cleanliness', 'Food and Drink',
    'In-flight Service', 'In-flight Wifi Service', 'In-flight Entertainment', 'Baggage Handling'
]

defaults = {
    'Gender': 1,
    'Customer Type': 1,
    'Type of Travel': 1,
    'Departure and Arrival Time Convenience': 3,
    'Ease of Online Booking': 3,
    'Online Boarding': 3,
    'Gate Location': 3,
    'On-board Service': 3,
    'Leg Room Service': 3,
    'Cleanliness': 3,
    'In-flight Wifi Service': 2,
    'In-flight Entertainment': 3
}

model_input = {}
for col in columns:
    if col in user_data:
        model_input[col] = user_data[col]
    else:
        model_input[col] = defaults.get(col, 3)  # default to 3 if not specified

new_passenger = pd.DataFrame([model_input])

predicted_class = xgb_model.predict(new_passenger)[0]
predicted_prob = xgb_model.predict_proba(new_passenger)[0][1]

print("Prediction:", "Satisfied" if predicted_class == 1 else "Neutral or Dissatisfied")
print(f"Probability of Satisfaction: {predicted_prob:.2f}")
```
